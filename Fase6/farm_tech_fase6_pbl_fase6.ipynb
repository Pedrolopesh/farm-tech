## 3. Avaliação Crítica dos Modelos
Abaixo, apresentamos a comparação entre as abordagens solicitadas: YOLOv5 (Transfer Learning adaptado) e CNN Customizada.

| Critério | YOLOv5 (Adaptável) | CNN (Treinada do Zero) |
| :--- | :--- | :--- |
| **Facilidade de Integração** | Complexa (exige anotação .txt) | Simples (exige apenas pastas) |
| **Precisão (mAP / Accuracy)** | Alta (Especializada em localização) | Alta (Focada apenas em rótulo) |
| **Tempo de Treinamento** | Moderado (~15-20 min / 60 épocas) | Rápido (~2 min / 10 épocas) |
| **Tempo de Inferência** | Tempo Real (Otimizado p/ GPU) | Rápido (Porém sem localização) |

### Análise de Achados:
1.  **YOLO 30 vs 60 Épocas:** Observou-se que o treinamento de 60 épocas reduziu significativamente a perda de localização (*box_loss*), permitindo que o modelo detectasse objetos parcialmente escondidos com mais confiança que o de 30 épocas.
2.  **YOLO vs CNN:** A CNN obteve uma acurácia de validação de ~90.9%, mas sua utilidade é limitada na fazenda, pois ela não indica a quantidade ou posição dos ativos, apenas "se" eles existem na foto. O YOLO provou ser a solução robusta para contagem e rastreio.

```python
import os
import shutil

# Criando estrutura de pastas para a CNN
os.makedirs('/content/cnn_dataset/vaca', exist_ok=True)
os.makedirs('/content/cnn_dataset/trator', exist_ok=True)

def preparar_para_cnn(origem_img, origem_lbl):
    for label_file in os.listdir(origem_lbl):
        if label_file.endswith('.txt'):
            with open(os.path.join(origem_lbl, label_file), 'r') as f:
                linha = f.readline()
                if linha:
                    classe = linha.split()[0] # 0 = vaca, 1 = trator
                    img_file = label_file.replace('.txt', '.jpg')

                    # Define o destino com base na classe do YOLO
                    destino = '/content/cnn_dataset/vaca' if classe == '0' else '/content/cnn_dataset/trator'

                    path_img = os.path.join(origem_img, img_file)
                    if os.path.exists(path_img):
                        shutil.copy(path_img, destino)

# Organizando fotos de treino e validação para a CNN
preparar_para_cnn('/content/dataset_final/train/images', '/content/dataset_final/train/labels')
preparar_para_cnn('/content/dataset_final/val/images', '/content/dataset_final/val/labels')

print("Fotos organizadas para a CNN com sucesso!")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Configuração do Dataset para a CNN
# 1. Configuração do Dataset para a CNN (Ajustado)
base_path = "/content/cnn_dataset" # Pasta nova organizada por subpastas
img_size = (128, 128)
batch_size = 16

# Agora o TensorFlow vai encontrar as classes 'vaca' e 'trator' automaticamente
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    base_path,
    label_mode='int',
    image_size=img_size,
    batch_size=batch_size,
    validation_split=0.2, # Separa 20% para validação
    subset="training",
    seed=123
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    base_path,
    label_mode='int',
    image_size=img_size,
    batch_size=batch_size,
    validation_split=0.2,
    subset="validation",
    seed=123
)

# ... resto do código da arquitetura CNN que passei antes ...

# 2. Arquitetura da Rede Neural (Conforme sugerido pela FIAP)
model_cnn = models.Sequential([
    layers.Rescaling(1./255, input_shape=(128, 128, 3)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='softmax') # 2 classes: vaca e trator
])

model_cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 3. Treinamento
history = model_cnn.fit(train_ds, validation_data=val_ds, epochs=10)
```
